```python

# Missing values present a major hurdle during analysis so we need to check for any available Nan Values
# below code represents the percentage of missing values in each variable
# from the output obtained we can see that there are no missing values present in the dataset
# below code tells use if its true or not true
print(df.isna().any())

# gives perecentage of the missing values
df.isnull().sum()/len(df)*100
```

```python
#since some of the columns like PiraniGauge,Element  contain unique values we can drop them as they will not help in our analysis
#converting the timestamp dtype to datetime will help us perform operations of datetime 
new_df = df.drop(columns=['PiraniGauge','Element'], axis=1)
new_df['Timestamp'] = pd.to_datetime(new_df['Timestamp'])

print(df.shape)
print(new_df.shape)
```

```python

#pair-wise correlation between columns
print(new_df.corr())

#check var
print(new_df.index)
```

```python
#since we have columns with only zero values we plan to drop those as seen in the describe cell
#from from stats above we can see that column HeliumPressure, MagnetTC and RGAChannel1-6 have 0 values which will not have much impact on our future analysis so we opt to drop those columns 

new_df1 = new_df.drop(columns=['HeliumPressure', 'MagnetTC', 'RGAChannel1' , 'RGAChannel2', 'RGAChannel3', 'RGAChannel4', 'RGAChannel4', 'RGAChannel5', 'RGAChannel6'])

#to dop the timestamp column so that we can plot the correlation between the variables 
new_df1 = new_df1.drop(columns=['Timestamp'])

print(new_df1.head(5))

#df1 = df.dropna(how='all', axis=1)
#print(df1.head(5))

'''
# Check if all values are zero in column 'colllo'
if (df[['RGAChannel1'][]]  == 0).all():
    print("All values in the column 'colllo' are Zero")
else:
    print("All values in the column 'colllo' are not Zero")
'''
```
