```python
#since some of the columns like PiraniGauge,Element  contain unique values we can drop them as they will not help in our analysis
#converting the timestamp dtype to datetime will help us perform operations of datetime 
new_df = df.drop(columns=['PiraniGauge','Element'], axis=1)
new_df['Timestamp'] = pd.to_datetime(new_df['Timestamp'])

print(df.shape)
print(new_df.shape)
```

```python
# set index as Timestamp
new_df1 = new_df.set_index(['Timestamp'])
new_df1.index = pd.to_datetime(new_df1.index)
#to dop the timestamp column so that we can plot the correlation between the variables 
#new_df1 = new_df.drop(columns=['Timestamp'])
```

```python
#demo pair-wise correlation between columns along with the heatmap for visualization
corr_matrix = new_df1.corr()

fig, ax = plt.subplots(figsize=(20,17)) 

sns.heatmap(corr_matrix, cmap='PuOr', annot=True, linewidths=.5)
```

```python
# give correlation betwwen the variables
corr_matrix = new_df1.corr().abs()
print(corr_matrix)
```

```python
# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
```

```python
# Find index of feature columns with correlation greater than 0.85
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]
print(to_drop)
```

```python
# Drop features 
new_df2 = new_df1.drop(new_df1[to_drop], axis=1)
new_df2
```

```python
#since we have columns with only zero values we plan to drop those as seen in the describe cell
#from stats above we can see that column HeliumPressure, MagnetTC and RGAChannel1-6 have 0 values which will not have much impact on our future analysis so we opt to drop those columns as well 

new_df3 = new_df2.drop(columns=['HeliumPressure', 'MagnetTC', 'RGAChannel1' , 'RGAChannel2', 'RGAChannel3', 'RGAChannel4', 'RGAChannel4', 'RGAChannel5', 'RGAChannel6'])
new_df3
```
